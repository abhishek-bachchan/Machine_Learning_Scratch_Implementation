```python
url='https://raw.githubusercontent.com/dipanjanS/practical-machine-learning-with-python/master/notebooks/Ch02_The_Python_ML_Ecosystem/simplemaps-worldcities-basic.csv'
df2=pd.read_csv(url,index_col=0,parse_dates=[0])
#we have sep parameter if seperator is not commoa
#we have header parameter, many cases header is present at top row=0 data will be from row 1 if we use header as row 0
#use names attributes if you want to explicitly provide name to each column[" "," "]
#use index_colto specify row labels to use if you set index_col=0 the first column of dataframe will become row label
#parse date is to make column which are date and time format to data and type otherwisse it will be treated as type 'object' not data and time
#https://honingds.com/blog/pandas-read_csv/  
#in above city column is used as row label as index_col=0
#read from above link if needed according to Demand
print(df2)
#you cannot print row label directly by df2['city']
print(df2.loc[['Zaranj']])
#above method is used to access any row using row labels df2.loc[['']]
print(df2['city_ascii'])
city_data=pd.read_csv(url)
#row label is generated automatically and top row is taken as column representative automatically
print(city_data)
```

```python
city_data.head(n=10)
#.head is used to access no of rown you want to show
#.tail is used to access few last rows
city_data.tail(n=10)
city_data[:7]
#examples shows slicing refers to slicing of rowsand all the other columns will end up in result
#for column slicing we use iloc
city_data.iloc[:5,:4]
#suppose we want to select city that have popu more than 10 mill and select column starts with letter 'l'
#city_data.columns give columns 
city_data[city_data['pop']>10000000][city_data.columns[pd.Series(city_data.columns).str.startswith('l')]]
city_greater_10mil=city_data[city_data['pop']>10000000]
city_greater_10mil.rename(columns={'pop':'population'},inplace=True)
print(city_greater_10mil)
city_greater_10mil.where(city_greater_10mil.population > 15000000)
```

```python
from numpy import nan
#Data operations
#value attribute allow to access raw values stored in dataframe and if they are all homogeneouswe can use numpy 
df3=pd.DataFrame(np.random.randn(8,3),columns=['A','B','C'])
print(df3.columns)
print(pd.Series(df3.columns))
#series covert into series such that oth index contain column 1 name 
print(pd.Series(df3.columns).str)
#column access can be done by above and below also
print(df3.A)
nparray=df3.values
nparray
#to fill missing values we will use fillna function
df3.iloc[4,2]=nan
print(df3)
df3.fillna(0)
```

```python
columns_numeric=['lat','lng','pop']
city_data[columns_numeric].mean()
print(city_data[columns_numeric].sum())
city_data['lat'].sum()
#similarly .count .median .quantile(0.8) quantile 0.5 is median
print(city_data[columns_numeric].quantile(0.75))
#we can get these value row wise using axis
print(city_data[columns_numeric].sum(axis=0))
print(city_data[columns_numeric].sum(axis=1))
#.describe can be used for each column
#concatening dataframes
city_data1=city_data.sample(3)
city_data2=city_data.sample(3)
city_data_combine=pd.concat([city_data1,city_data2])
print(city_data_combine)
```
